```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('tips.csv')
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (9,5))
sns.countplot(data = df , x = 'tip')
plt.title('Distribution of tip')
plt.xlabel('tip')
plt.ylabel('Count')
plt.show()
```

```python
df['tip'].value_counts
```

```python
plt.style.use('default')
plt.bar(df.groupby('smoker')['tip'].mean().index, df.groupby('smoker')['tip'].mean().values)
plt.title('smoker average')
plt.xlabel('smoker')
plt.ylabel('average')
plt.show()
```

```python
p=df.groupby('smoker')['tip'].mean()
p
```

```python
plt.style.use('default')
plt.bar(df.groupby('smoker')['total_bill'].mean().index, df.groupby('smoker')['total_bill'].mean().values)
plt.title('smoker average')
plt.xlabel('smoker')
plt.ylabel('average')
```

```python
p=df.groupby('smoker')['total_bill'].mean().index
p
```

```python
plt.pie(df.groupby('smoker')['total_bill'].mean().values,autopct='%.2f', labels=df.groupby('smoker')['total_bill'].mean().index)
plt.title('total bill piechart')
plt.show()
```

```python
p=df.groupby('smoker')['total_bill'].mean().values
p
```

```python
plt.pie(df.groupby('day')['total_bill'].sum(),autopct='%.2f',labels=df.groupby('day')['total_bill'].sum().index)
plt.title('all day')
plt.show()
```

```python
a=df.groupby('day')['total_bill'].sum()
a
```

```python
plt.bar(df.groupby('day')['tip'].sum().index,df.groupby('day')['tip'].sum().values)
plt.title('per day total tip')
plt.xlabel('day')
plt.ylabel('values')
plt.show()
```

```python
s=df.groupby('day')['tip'].sum().index
s
```

```python
z=df.groupby('day')['tip'].sum().values
z
```

```python
plt.bar(df.groupby('sex')['tip'].mean().index, df.groupby('sex')['tip'].mean().values)
plt.title('tips given')
plt.xlabel('gender')
plt.ylabel('tip avg')
plt.show()
```
