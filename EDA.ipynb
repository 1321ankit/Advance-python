# Bivariate Analysis

```python
#1. scatterplot(Numerical column - Numerical column)
```

```python
#scatterplot is used in numerical data and barplot is used for categorical data.
```

```python
tips= sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x=tips['total_bill'] , 
               y=tips['tip'])
```

```python
# hue, style, color ,size ==> use for comparision between columns.
```

```python
sns.scatterplot(x='total_bill',
                y='tip', data=tips, hue=tips['sex'])
```

# Multivariate Analysis

```python
sns.scatterplot(x='total_bill',
               y='tip', data=tips,
               hue=tips['sex'], style=tips['smoker'])
```

```python
# HeatMap(Categorical-categorical)
p=pd.crosstab(tips['day'],tips['time'])
p
```

```python
sns.heatmap(p)
```
